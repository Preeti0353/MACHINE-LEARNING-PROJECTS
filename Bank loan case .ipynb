# Data Pre Processing (Cleaning)

```python

def find_missing_percentage(df):
    missing_percentage = df.isnull().sum() / len(df) * 100
    missing_percentage_df = missing_percentage.sort_values(ascending=False)
    return missing_percentage_df
```

```python
find_missing_percentage(df)
```

```python
#Dropping columns with more than 40% null values from application dataset
#49 columns are there with more than 40% null values

limitPer = len(df) * 0.60
df = df.dropna(thresh=limitPer,axis=1)
```

```python
find_missing_percentage(df)
```

```python
#'Occupation Type' has 31% missing values and since this might help us in analysis, let's not drop this.
#Also, since 31% are null, we cannot impute as the analysis might get biased so, let's replace the null values with 'UNKNOWN'

df['OCCUPATION_TYPE'] = df['OCCUPATION_TYPE'].fillna('na')
```

```python
df.OCCUPATION_TYPE.value_counts()
```

```python
# Function to replace null values of numerical columns with median
#Replacing numberical columns with 20% or less null values with median as many of them seem to have outliers and 
#cannot be replaced with mean since replacing with mean can lead to bias in our analysis.
columns_with_missing = ['EXT_SOURCE_3','AMT_REQ_CREDIT_BUREAU_YEAR','AMT_REQ_CREDIT_BUREAU_QRT', 'AMT_REQ_CREDIT_BUREAU_MON', 
                 'AMT_REQ_CREDIT_BUREAU_WEEK', 'AMT_REQ_CREDIT_BUREAU_DAY', 'AMT_REQ_CREDIT_BUREAU_HOUR','EXT_SOURCE_2',
                 'AMT_GOODS_PRICE','AMT_ANNUITY','OBS_30_CNT_SOCIAL_CIRCLE', 'DEF_30_CNT_SOCIAL_CIRCLE', 
                 'OBS_60_CNT_SOCIAL_CIRCLE', 'DEF_60_CNT_SOCIAL_CIRCLE','CNT_FAM_MEMBERS','DAYS_LAST_PHONE_CHANGE']
# Fill missing values with median
for column in columns_with_missing:
    median_value = df[column].median()
    df[column].fillna(median_value, inplace=True)
```

```python
#Replacing the missing values in Categorical variables NAME_TYPE_SUITE with the MODE value 'Unaccompanied'
df['NAME_TYPE_SUITE'].fillna(df.NAME_TYPE_SUITE.mode()[0], inplace=True)
```

```python
# Identify columns with missing values
columns_with_missing = ['NAME_TYPE_SUITE']
# Fill missing values with mode
for column in columns_with_missing:
    mode_value = df[column].mode().values[0]
    df[column].fillna(mode_value, inplace=True)
```

```python
find_missing_percentage(df)
```

```python
#Checking Data Types
#Following columns are in float but since they are discrete variables, converting them to integer
df['DAYS_REGISTRATION'] = df['DAYS_REGISTRATION'].astype(int)
df['CNT_FAM_MEMBERS'] = df['CNT_FAM_MEMBERS'].astype(int)
df['AMT_REQ_CREDIT_BUREAU_HOUR'] = df['AMT_REQ_CREDIT_BUREAU_HOUR'].astype(int)
df['AMT_REQ_CREDIT_BUREAU_YEAR'] = df['AMT_REQ_CREDIT_BUREAU_YEAR'].astype(int)
df['DEF_60_CNT_SOCIAL_CIRCLE'] = df['DEF_60_CNT_SOCIAL_CIRCLE'].astype(int)
df['OBS_60_CNT_SOCIAL_CIRCLE'] = df['OBS_60_CNT_SOCIAL_CIRCLE'].astype(int)
df['DAYS_LAST_PHONE_CHANGE'] = df['DAYS_LAST_PHONE_CHANGE'].astype(int)
df['AMT_REQ_CREDIT_BUREAU_QRT'] = df['AMT_REQ_CREDIT_BUREAU_QRT'].astype(int)
df['DEF_30_CNT_SOCIAL_CIRCLE'] = df['DEF_30_CNT_SOCIAL_CIRCLE'].astype(int)
df['AMT_REQ_CREDIT_BUREAU_MON'] = df['AMT_REQ_CREDIT_BUREAU_MON'].astype(int)
df['OBS_30_CNT_SOCIAL_CIRCLE'] = df['OBS_30_CNT_SOCIAL_CIRCLE'].astype(int)
df['AMT_REQ_CREDIT_BUREAU_WEEK'] = df['AMT_REQ_CREDIT_BUREAU_WEEK'].astype(int)
df['AMT_REQ_CREDIT_BUREAU_DAY'] = df['AMT_REQ_CREDIT_BUREAU_DAY'].astype(int)
```
